```python
def make_multiframe_from_wrong(pdbfile):
    with open(pdbfile, 'r') as f:
        pdbcrap = f.read()
        
    for i, item in enumerate(pdbcrap.split('END\n')):
        with open("{}_frame{}.pdb".format(pdbfile, i), 'w') as f:
            f.write(item)
    
    return pdbcrap
```
